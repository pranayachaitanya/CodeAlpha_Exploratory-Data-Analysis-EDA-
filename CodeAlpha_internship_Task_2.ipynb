```python
from scipy.stats import ttest_ind
import pandas as pd

# Ensure 'discount_percentage' is numeric
df['discount_percentage_cleaned'] = df['discount_percentage'].astype(str).str.replace('%', '').astype(float)

# Fill any NaN values that might have resulted from cleaning in discount_percentage_cleaned
df['discount_percentage_cleaned'] = df['discount_percentage_cleaned'].fillna(df['discount_percentage_cleaned'].mean())

# Define groups based on discount percentage (e.g., above and below median discount)
median_discount = df['discount_percentage_cleaned'].median()

high_discount_products = df[df['discount_percentage_cleaned'] > median_discount]
low_discount_products = df[df['discount_percentage_cleaned'] <= median_discount]



# Extract ratings for each group
high_discount_ratings = high_discount_products['rating_cleaned']
low_discount_ratings = low_discount_products['rating_cleaned']

# Perform independent samples t-test
t_stat, p_value = ttest_ind(high_discount_ratings, low_discount_ratings, nan_policy='omit')

print(f"Hypothesis: Is there a significant difference in average ratings between products with above-median discount percentages and products with below-median discount percentages?")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Conclusion: The p-value is less than 0.05, so we reject the null hypothesis. There is a statistically significant difference in average ratings between the two groups.")
else:
    print("Conclusion: The p-value is not less than 0.05, so we fail to reject the null hypothesis. There is no statistically significant difference in average ratings between the two groups.")
```
